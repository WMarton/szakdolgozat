```python
nltk.download('words')
nltk.download('punkt')
from nltk.tokenize import word_tokenize
from nltk.corpus import words
in_path = 'filtered'
out_path = 'cleaned'
in_files = [f for f in listdir(in_path) if isfile(join(in_path, f))]
to_remove = ["tank", "javelin", "times"]
english = set(words.words()).difference(to_remove) 
for file_name in in_files:
  article = open(join(in_path,file_name), "r").read()
  proc = []
  doc = word_tokenize(article)
  for token in doc:
    if token not in english:
      proc.append(str(token))
    txt = ' '.join(proc)
  with codecs.open(join(out_path, file_name), 'w', 'utf-8') as out_file:
      out_file.write(txt)
    
```

```python
import operator
from collections import Counter

in_path = 'cleaned' 
in_files = [f for f in listdir(in_path) if isfile(join(in_path, f))]

txts = []

for f in in_files:
    try:
        txt = codecs.open(join(in_path, f), 'r', 'utf-8').read()
        txt = txt.strip().split()
        txt = [w for w in txt if len(w) > 3]
        txts.append(txt)
    except Exception as e:
        print(f, e)
        
wordlist = list(itertools.chain(*txts))
wfreq = Counter(wordlist)
sorted_wfreq = sorted(wfreq.items(),
                      key=operator.itemgetter(1),
                      reverse=True)

with codecs.open('wfreq.tsv', 'w', 'utf-8') as f:
    for e in sorted_wfreq:
        o = e[0] + '\t' + str(e[1]) + '\n'
        f.write(o)

wd_docfreq = {}
for d in txts:
    d = list(set(d))
    for w in d:
        if w not in wd_docfreq.keys():
            wd_docfreq[w] = 1
        else:
            wd_docfreq[w] += 1

sorted_docfreq = sorted(wd_docfreq.items(),
                        key=operator.itemgetter(1),
                        reverse=True)

with codecs.open('docfreq.tsv', 'w', 'utf-8') as f:
    for e in sorted_docfreq:
        o = e[0] + '\t' + str((e[1]/len(txts))) + '\n'
        f.write(o)
```

```python
content_words = []
of = codecs.open('contentwords.tsv', 'w', 'utf-8')
with codecs.open('docfreq.tsv', 'r', 'utf-8') as input_file:
    for l in input_file:
        wd, dfreq = l.strip().split('\t')
        dfreq = float(dfreq)
        if dfreq > 0.005:
            content_words.append(wd)
            o = wd + '\t' + str(dfreq) + '\n'
            of.write(o)
of.close()
```

```python
content_words = []
with codecs.open('contentwords.tsv', 'r', 'utf-8') as input_file:
    for l in input_file:
        cw, wfreq = l.strip().split('\t')
        content_words.append(cw)

in_path = 'cleaned'
out_path = 'finalized'
files = sorted([f for f in listdir(in_path) if isfile(join(in_path, f))])
file = 0

for f in files:
    with codecs.open(join(out_path, files[file]), 'w', 'utf-8') as out_file:
        try:
            txt = codecs.open(join(in_path, f), 'r', 'utf-8').read()
            txt = txt.strip().split()
            for cw in txt:
                if cw in content_words:
                    out_file.write(cw + " ")
        except Exception as e:
            print(files[file], e)
    file += 1
```
